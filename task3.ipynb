# Task III — Open Task

## Why symmetry-aware quantum circuits are the right direction for HEP

After working on Task II (quark/gluon classification with GCNs and GATs) and Task VII (equivariant QNNs), I want to share some thoughts on where I think quantum machine learning can actually be useful in high energy physics — and where the hype might be getting ahead of the reality.

### The expressibility problem

One thing that struck me when implementing the standard QNN in Task VII is how easy it is to build a variational circuit that looks expressive on paper but gets stuck in practice. With random initialization on a generic ansatz, the gradient landscape becomes exponentially flat as you add qubits — this is the barren plateau problem (McClean et al., 2018). In my experience training the 1-qubit standard QNN, convergence was already somewhat sensitive to initialization. Scaling this naively to tens of qubits for a real HEP problem seems problematic.

This is why I think the equivariant approach from Meyer et al. (2022) is genuinely promising and not just a theoretical curiosity. By constraining the circuit to respect the symmetry group of the problem, you're effectively reducing the search space of the optimizer. In Task VII, the Z₂×Z₂ equivariant QNN achieved comparable or better accuracy than the standard QNN with fewer parameters — and that's on a toy 2D problem. The gains should be even more significant on higher-dimensional HEP data where the unconstrained parameter space is enormous.

### Connecting this to jet physics

What excites me about the equivariant QNN project for LHC analysis is that particle physics is full of exact symmetries that we can exploit. Jets have a natural permutation symmetry (the ordering of constituents is arbitrary), and there are Lorentz symmetries in the kinematic variables. In the classical world, architectures like ParticleNet and LorentzNet already leverage some of these symmetries through their design. The question is whether quantum circuits can do the same more efficiently.

From what I've seen in Task II, the attention mechanism in GAT learns to focus on the most informative particles — but it does so in a brute-force way, with no symmetry guarantees. An equivariant quantum circuit could potentially achieve similar selectivity while being constrained to respect the physics by construction. Whether this leads to a practical advantage on current (noisy) quantum hardware is still an open question, but at the theoretical level the argument is compelling.

### A note on PennyLane

On the software side, I used PennyLane for Tasks I and VII and found it very well suited for this kind of work. The ability to define quantum circuits as differentiable programs and plug them into a standard gradient-based training loop (with `qml.AdamOptimizer`) makes the gap between classical and quantum ML feel surprisingly small from a development perspective. The `qml.Hermitian` observable was particularly useful for the equivariant circuit where we needed a custom measurement operator. I think PennyLane's approach of treating quantum circuits as just another differentiable layer is the right abstraction for hybrid quantum-classical models.

### What I'd like to explore

If selected for the equivariant QNN project, I'd like to push beyond the Z₂×Z₂ example and investigate how to encode permutation symmetry (the symmetric group $S_n$) into quantum circuits for jet-level classification. The challenge is that $S_n$ grows factorially, so a naive approach won't work — but there's recent theoretical work on efficient equivariant circuit constructions that I think could be adapted to the jet classification setting. I'm also curious about whether equivariant circuits are more robust to hardware noise, since the constrained parameter space might make them less sensitive to gate errors. This would be important for any eventual deployment on real quantum hardware.
